```python
import numpy as np
import myenvironment as myenv
import random
import policy

game = myenv.environment_easy21()
pi = policy.policies(None)


        

class MCmodel:
    def __init__(self):
        self.theta = np.random.randn(12)/np.sqrt(12)
        train_set = []
        for n in range(22):
            ones = 22 - n
            for o in range(ones):
                for action in range(2):
                    for d in range(10):
                        d = d+1
                        train_set.append([o,n,d,action])
        self.train_set = train_set
        
    
    
    def action_value2list(self,state,action):
        state_action = state[0]
        state_action.append(int(state[1]))
        state_action.append(int(action))
        return state_action
    
    def list2action_value(self,s_a):
        p = s_a[:-2]
        d = s_a[-2]
        a = s_a[-1]
        return [p,d],a

    def count_sa(self,count_dic,s_a):
        if str(s_a) in count_dic:
            count_dic[str(s_a)] += 1
        else:
            count_dic[str(s_a)] = 1
        return count_dic
    
    def rad_train(self,epoch):
        count_dic={}
        Q = {}
        
        for epoches in range(epoch):
            for s_a in self.train_set:
                count_dic = self.count_sa(count_dic,s_a)
                state,action = self.list2action_value(s_a)
                r = game.play(state,action,pi.random_policy)
                #print('state,action,r=',state,action,r)
                if str(s_a[:-1]) in Q.keys():
                    if str(s_a[-1]) in Q[str(s_a[:-1])].keys():
                        #print(Q[str(s_a[:-1])])
                        previous_sum = (count_dic[str(s_a)]-1)*Q[str(s_a[:-1])][str(s_a[-1])]
                        Q[str(s_a[:-1])][str(s_a[-1])] = (previous_sum+r) /count_dic[str(s_a)]
                    else:
                        Q[str(s_a[:-1])][str(s_a[-1])] = r
                else:
                    Q[str(s_a[:-1])] = {}
                    Q[str(s_a[:-1])][str(s_a[-1])] = r
                    
        return Q

    def further_train(self,epoch,Q_input):
        count_dic={}
        Q = {}
        pi_Q = policy.policies(Q_input)
        
        for epoches in range(epoch):
            for s_a in self.train_set:
                count_dic = self.count_sa(count_dic,s_a)
                state,action = self.list2action_value(s_a)
                r = game.play(state,action,pi_Q.max_q_action)
                #print('state,action,r=',state,action,r)
                if str(s_a[:-1]) in Q.keys():
                    if str(s_a[-1]) in Q[str(s_a[:-1])].keys():
                        #print(Q[str(s_a[:-1])])
                        previous_sum = (count_dic[str(s_a)]-1)*Q[str(s_a[:-1])][str(s_a[-1])]
                        Q[str(s_a[:-1])][str(s_a[-1])] = (previous_sum+r) /count_dic[str(s_a)]
                    else:
                        Q[str(s_a[:-1])][str(s_a[-1])] = r
                else:
                    Q[str(s_a[:-1])] = {}
                    Q[str(s_a[:-1])][str(s_a[-1])] = r
                    
        return Q
```
