```python
import numpy as np
import random
class environment_easy21:
    
    def __init__(self):
        a = list(np.arange(1,11))
        for n in 10*np.ones(3):
            a.append(int(n))
        self.cards = a
        self.action = [0,1] #0 stands for stick, 1 stands for hit
    
    #convert card number to the count of each card appears   
    def to_10_form(self,player):
         init_state = np.zeros((10,),dtype = int)
         for p in player:
             init_state[p-1] += 1
         init_state = list(init_state)
         return init_state
        
    def to_update_form(self,state_10):
        state = [state_10[0]]
        card_form = self.to_card_form(state_10)
        sum_s = sum(card_form)-state_10[0]
        state = [state_10[0],sum_s]
        return state
        
        
     #convert back to card number form
    def to_card_form(self,list):
         player = []
         n = 0
         for k in list:
             k = int(k)
             n += 1
             if k > 0:
                 for number in n*np.ones(k):
                     player.append(int(number))
         return player
           
    #give the player and the player one card
    #the output is in update form
    def initial_game(self):
        s1 = random.choice(self.cards)
        s2 = random.choice(self.cards)
        player = [s1,s2]
        player = self.to_10_form(player)
        player = self.to_update_form(player)
        dealer = random.choice(self.cards)
        return player,dealer
    
    
    def step(self,state,action):
        #state is in the [update_form, d]
        p, d = state
        sum_d = d
        terminate = False
        r = 0
        
        if action == 1:
            add_ = random.choice(self.cards)
            if add_ == 1:
                p[0] += 1
            else:
                p[1] += add_
        else:
            terminate = True
        
        #print(p)
        
        def sum_player(state):
            sum_p = [sum(p)]
            if state[0] != 0:
                for n in range(state[0]):
                    sum_p.append((n+1)*10+sum(state))
                #print(sum_p)
                entry = 0
                for i in sum_p:
                    if i <= 21 and i > entry:
                        entry = i
                if entry > 0:
                    sum_p = entry
                if np.all(np.array(sum_p)>21):
                    sum_p = 22
            else:
                sum_p = sum(p)
            return sum_p
        
        sum_p = sum_player(p)
        #print('sum_p=',sum_p)
        
        if sum_p > 21:
            terminate = True
            r = -1
        elif sum_p <= 21 and terminate == True:
                #dealer's cards
                while sum_d<17:
                    add_d = random.choice(self.cards)
                    if add_d ==1:
                        if sum_d + 11 < 22:
                            sum_d = sum_d + 11
                        else:
                            sum_d = sum_d + 1
                    else:
                        sum_d = sum_d + add_d
                        
                if sum_d > 21 or sum_d < sum_p:
                    r = 1
                    terminate = True
                elif sum_d == sum_p:
                    r = 0
                    terminate = True
                elif sum_d > sum_p: 
                    r = -1
                    terminate = True
        
        state_new = [p,sum_d]
        return state_new, r, terminate
    #the state_new's p is of update form            
        
    def play(self,state,action,policy):
        state_new,r,terminate = self.step(state,action)
        #print(state_new,r,terminate)
        while terminate == False:
            state = state_new
            action = policy(state)
            #print('action=',action)
            state_new, r, terminate = self.step(state,action)
            #print(state_new,r,terminate)
        return r

#Y = environment_easy21()


#y = Y.play([[0,10], 10],1)
#print(y)
```
